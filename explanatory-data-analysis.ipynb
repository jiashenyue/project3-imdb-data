## Concatenate dataframes

```python
# read the rating file to obtain the tconst of each movie for analysis
df_00 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_00.head()
```

```python
df_01 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
df_01.head()
```

```python
# concatenate data
frames = [df_00, df_01]
df_merge = pd.concat(frames)
display(df_merge.head(), df_merge.tail())
```

```python
# clean dataframe
# rmmove imdb_id == 0
df = df_merge[df_merge['imdb_id'] != '0']
df.head()
```

```python
# setup a few static parameters for plotting
col_pal = sns.color_palette('ch:1, r=1.45, l=.55, d=.395')
sns.color_palette('ch:1, r=1.45, l=.55, d=.395')
```

#### Box plot

```python
# Create a function to create a KDE Plot
def box_plot(df, column_name,
             fs=(8,4), file_name='',
             tit_lab='', x_lab='', y_lab='', 
             fmt='',
             hza='center', rot=0):    
    # Variables
    feature = df[column_name]
    mean = feature.mean()
    median = feature.median()
    # Plot
    fig, ax = plt.subplots(nrows=1, figsize=fs, facecolor='w')
    sns.boxplot(data=df, x=column_name, width=.35, color=col_pal[4], ax=ax,
                medianprops={'color':'k', 'linewidth':1.5},
                whiskerprops={'color':'k', 'linewidth':1.5},
                boxprops={'facecolor':col_pal[4], 
                          'edgecolor':'k', 'linewidth':1.5},
                capprops={'color':'k', 'linewidth':1.5},
                flierprops={'marker':'o', 'markersize':8, 
                            'markerfacecolor':col_pal[4], 
                            'markeredgecolor':'k'}); 
    # Title and labels
    plt.title(tit_lab, fontsize = 16, weight='bold')
    plt.xlabel(x_lab, fontsize = 12, weight='bold')
    plt.ylabel('', fontsize = 12, weight='bold')
    # Ticks
    plt.xticks(fontsize=10, weight='bold')
    plt.yticks(fontsize=10, weight='bold')
    # Face 
    ax.set_facecolor('w')
    # Spines
    for axis in ['top','bottom','left','right']:
        ax.spines[axis].set_linewidth(2)
    # Vertical lines
    ax.axvline(mean, color = 'magenta', linewidth=2, 
               label=f'Mean = {mean:,.2f}')
    ax.axvline(median, ls='dotted', color = 'darkmagenta',  linewidth=2, 
               label=f'Median = {median:,.2f}')
    # Grid
    plt.grid(axis = 'x', lw=.5, c='k', alpha=.25)
    # Display
    ax.legend();
    plt.tight_layout()
    plt.show;
```

#### Histogram

```python
# a function to create histogram
# Create a function to create a Histogram
def hist_plot(df, column_name, bin_count='auto',
              fs=(8,4), file_name='',
              tit_lab='', x_lab='', y_lab='', 
              hza='center', rot=0):
    # Variables
    feature = df[column_name]
    mean = feature.mean()
    median = feature.median()
    
    # Plot
    fig, ax = plt.subplots(figsize=fs, facecolor='w')
    # fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (8,4));
    sns.histplot(data=df, x=column_name, 
                 color=col_pal[4], linewidth=1.25, bins=bin_count)
    
    # Title and labels
    plt.title(tit_lab, fontsize = 16, weight='bold')
    plt.xlabel(x_lab, fontsize = 12, weight='bold')
    plt.ylabel(y_lab, fontsize = 12, weight='bold')
    
    # Ticks
    plt.xticks(fontsize=10, weight='bold')
    plt.yticks(fontsize=10, weight='bold')
    # Face 
    ax.set_facecolor('w')

    # Vertical lines
    ax.axvline(mean, color = 'magenta', linewidth=2, 
               label=f'Mean = {mean:,.2f}')
    ax.axvline(median, ls='dotted', color = 'darkmagenta',  linewidth=2, 
               label=f'Median = {median:,.2f}')
    # Grid
    plt.grid(lw=.5, c='k', alpha=.25)
    # Display
    ax.legend();
    plt.tight_layout()
    plt.show;
```
